```python
merged_df=train_df.merge(holiday_df, on= 'date')
merged_df=merged_df.merge(oil_df, on='date', how='left')
merged_df=merged_df.merge(store_df, on='store_nbr', how='left')
merged_df=merged_df.merge(trans_df, on=['date', 'store_nbr'], how='left')
merged_df
```

```python
merged_df=merged_df.rename(columns = {"type_x" : "holiday_type", "type_y" : "store_type"})
```

```python
pd.set_option('display.max_columns', 50)
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df['year'] = merged_df['date'].dt.year
merged_df['month'] = merged_df['date'].dt.month
merged_df['week'] = merged_df['date'].dt.isocalendar().week
merged_df['quarter'] = merged_df['date'].dt.quarter
merged_df['day_of_week'] = merged_df['date'].dt.day_name()
merged_df[:2]
```
